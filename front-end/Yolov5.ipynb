```python
import os
import shutil
import random

# Specify the path to the extracted images
data_path = './extracted_files/model/Data'

# Create directories for train, val, and test
train_dir = './data/train'
val_dir = './data/val'
test_dir = './data/test'

os.makedirs(train_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)
```

```python
#Split the data in to training testing and valadating parts
# Get a list of all images
all_images = os.listdir(data_path)

# Shuffle the images
random.shuffle(all_images)

# Calculate the split indices
total_images = len(all_images)
train_size = int(total_images * 0.7)
val_size = int(total_images * 0.15)
test_size = total_images - train_size - val_size

# Split the images into train, val, and test sets
train_images = all_images[:train_size]
val_images = all_images[train_size:train_size + val_size]
test_images = all_images[train_size + val_size:]

# Move images to their respective directories
for img in train_images:
    shutil.move(os.path.join(data_path, img), os.path.join(train_dir, img))

for img in val_images:
    shutil.move(os.path.join(data_path, img), os.path.join(val_dir, img))

for img in test_images:
    shutil.move(os.path.join(data_path, img), os.path.join(test_dir, img))

print(f"Training images: {len(train_images)}")
print(f"Validation images: {len(val_images)}")
print(f"Testing images: {len(test_images)}")
```

```python
# Create the data.yaml file
data_yaml_content = """
train: ./data/train  # Path to training images
val: ./data/val      # Path to validation images
test: ./data/test    # Path to testing images (optional)
nc: 2                # Number of classes (adjust as needed)
names: ['class1', 'class2']  # Replace with your class names
"""

with open('data.yaml', 'w') as f:
    f.write(data_yaml_content)
```

```python
import os
import xml.etree.ElementTree as ET

# Define the path to your XML files and where to save YOLO labels
xml_dir = '/content/yolov5/extracted_files/model/eyegaze/Annotations/'  # Correct path to XML directory
yolo_dir = '/content/yolov5/data/labels/'  # Change to your YOLO labels directory

# Create the YOLO labels directory if it doesn't exist
os.makedirs(yolo_dir, exist_ok=True)

# Function to convert a single XML file to YOLO format
def convert_xml_to_yolo(xml_file):
    tree = ET.parse(xml_file)
    root = tree.getroot()

    # Get the image dimensions
    width = int(root.find('size/width').text)
    height = int(root.find('size/height').text)

    yolo_lines = []

    # Iterate through each object in the XML
    for obj in root.findall('object'):
        class_name = obj.find('name').text
        if class_name in class_mapping:
            class_id = class_mapping[class_name]
        else:
            print(f"Warning: Class '{class_name}' not found in class_mapping.")
            continue  # Skip this object if class not found
        bbox = obj.find('bndbox')
        xmin = int(bbox.find('xmin').text)
        ymin = int(bbox.find('ymin').text)
        xmax = int(bbox.find('xmax').text)
        ymax = int(bbox.find('ymax').text)

        # Convert bounding box to YOLO format
        x_center = (xmin + xmax) / 2 / width
        y_center = (ymin + ymax) / 2 / height
        bbox_width = (xmax - xmin) / width
        bbox_height = (ymax - ymin) / height

        yolo_lines.append(f"{class_id} {x_center} {y_center} {bbox_width} {bbox_height}")

    return yolo_lines


# Class mapping from names to IDs (adjust this according to your dataset)
class_mapping = {
    'class_name1': 0,  # Replace 'class_name1' with actual class name and update IDs accordingly
    'class_name2': 1,
    # Add more classes as needed
}

# Iterate through all XML files in the directory
for xml_file in os.listdir(xml_dir):
    if xml_file.endswith('.xml'):
        xml_path = os.path.join(xml_dir, xml_file)
        yolo_lines = convert_xml_to_yolo(xml_path)

        # Create corresponding YOLO label file
        yolo_file_name = xml_file.replace('.xml', '.txt')
        yolo_file_path = os.path.join(yolo_dir, yolo_file_name)

        with open(yolo_file_path, 'w') as yolo_file:
            yolo_file.write('\n'.join(yolo_lines))

print("Conversion complete!")
```
